```python
import pandas as pd
```

```python
import sqlalchemy as sa
```

```python
import requests
```

```python
source='InciWeb,EO'
status='open'
days=20
```

```python
api_url = 'https://eonet.gsfc.nasa.gov/api/v2.1/events?source={}&status={}&days={}'.format(source,status,days)
```

```python
print(api_url)
```

```python
api_response=requests.get(api_url)
```

```python
events = api_response.json()
events
```

```python
df = pd.DataFrame(events['events'])
```

```python
df
```

```python
df = df.explode('geometries')
```

```python
df
```

```python
df_geometries = df['geometries'].apply(pd.Series)

# Now, let's concatenate this data back to our original dataframe
df = pd.concat([df.drop(['geometries'], axis=1), df_geometries], axis=1)
```

```python
df
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
relevant_col = ['title','date','coordinates']
```

```python
df.info()
```

```python
final_df = df[relevant_col]
```

```python
final_df
```

```python
final_df[['longitude', 'latitude']] = pd.DataFrame(df['coordinates'].tolist(), index=df.index)
final_df = final_df.drop(columns=['coordinates'])
```
